```python
import pandas as pd
```

```python
Flight = pd.read_excel("/content/drive/My Drive/FYP2/FYP2_FlightDataset_Final.xlsx")
Weather = pd.read_excel("/content/drive/My Drive/FYP2/FYP2_WeatherDataset_Final.xlsx")
FootTraffic = pd.read_csv("/content/drive/My Drive/FYP2/FYP2_FootTrafficDataset_Final.csv")
```

```python
ArrivalWeather = Weather.copy()
DepartureWeather = Weather.copy()
ArrivalWeather.columns = 'Arrival_' + ArrivalWeather.columns.values
DepartureWeather.columns = 'Departure_' + DepartureWeather.columns.values

ArrivalFootTraffic = FootTraffic.copy()
DepartureFootTraffic = FootTraffic.copy()
ArrivalFootTraffic.columns = 'Arrival_' + ArrivalFootTraffic.columns.values
DepartureFootTraffic.columns = 'Departure_' + DepartureFootTraffic.columns.values
```

### Weather

```python
#Departure
merge_df = pd.merge(Flight, DepartureWeather, how="left", left_on=['dep_iata','dep_Date_utc','dep_Hour_utc'], right_on=['Departure_Airport', 'Departure_Date','Departure_Hour']);
#Arrival
merge_df = pd.merge(merge_df, ArrivalWeather, how="left", left_on=['arr_iata','arr_Date_utc','arr_Hour_utc'], right_on=['Arrival_Airport', 'Arrival_Date','Arrival_Hour']);
```

### Foot Traffic

```python
#Departure
merge_df = pd.merge(merge_df, DepartureFootTraffic, how="left", left_on=['dep_iata','dep_Day_local','dep_Hour_local'], right_on=['Departure_Airport', 'Departure_Day','Departure_Hour']);
#Arrival
merge_df = pd.merge(merge_df, ArrivalFootTraffic, how="left", left_on=['arr_iata','arr_Day_local','arr_Hour_local'], right_on=['Arrival_Airport', 'Arrival_Day','Arrival_Hour']);
```

```python
merge_df = merge_df.drop(columns=['dep_iata','arr_iata','dep_Hour_utc','dep_Day_local','dep_Hour_local',
                                  'arr_Hour_utc','arr_Day_local','arr_Hour_local',
                                  'Departure_Airport_x','Departure_Date','Departure_Hour_x',
                                  'Arrival_Airport_x','Arrival_Date','Arrival_Hour_x',
                                  'Departure_Airport_y','Departure_Day','Departure_Hour_y',
                                  'Arrival_Airport_y','Arrival_Day','Arrival_Hour_y'])
```

```python
from google.colab import files

merge_df.to_excel('FYP2_Data_Final.xlsx', encoding = 'utf-8-sig',index=False)
files.download('FYP2_Data_Final.xlsx')
```
